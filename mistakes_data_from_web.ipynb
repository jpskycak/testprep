```python
import requests
from bs4 import BeautifulSoup
from itertools import compress

def unlist(l):
    return [item for sublist in l for item in sublist]

def cleanup(text,oldnew={'\xe2\x80\x99': "'", '\xe2\x80\x9c': '\"', '\xe2\x80\x9d':'\"', '\xe2\x80\xa6': '...', '\xc2\xa0':''}):
    for old, new in oldnew.iteritems():
        text = text.replace(old,new)
    return text

def lsathacks_RC(test_number=77, question_number=5):
    url = 'https://lsathacks.com/explanations/lsat-preptest-'+str(test_number)+'/reading-comprehension/q-'+str(question_number)+'/'
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text)
    
    d = {'test_number': test_number,
        'section_type': 'Reading Comprehension',
        'question_number': question_number}
    
    s1 = [x.contents for x in soup.find_all('span', class_='s1')]
    d['question_intro'] = cleanup(s1[0][1].encode('utf-8'))
    
    s2 = [x.contents for x in soup.find_all('span', class_='s2')]
    s2 = [cleanup(''.join([y.encode('utf-8') for y in x])).strip() for x in s2]
    d['discussion'] = ''.join(s2[:s2.index('___________')])
    d['explanations'] = s2[s2.index('___________')+1:]
        
    return d


def lsathacks_LR(test_number=77, order=1, question_number=5):
    url = 'https://lsathacks.com/explanations/lsat-preptest-'+str(test_number)+'/logical-reasoning-'+str(order)+'/q-'+str(question_number)+'/'
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36'}
    r = requests.get(url, headers=headers)
    soup = BeautifulSoup(r.text)
    
    d = {'test_number': test_number,
        'section_type': 'Logical Reasoning',
        'order': order,
        'question_number': question_number}
    
    s1 = [x.contents for x in soup.find_all('span', class_='s1')]
    d['question_intro'] = cleanup(s1[0][1].encode('utf-8'))
    
    s2 = [x.contents for x in soup.find_all('span', class_='s2')]
    s2 = [cleanup(''.join([y.encode('utf-8') for y in x])).strip() for x in s2]
    d['question_type'] = s2[0]
    d['conclusion'] = s2[1]
    d['reasoning'] = s2[2]
    d['analysis'] = ''.join(s2[3:s2.index('___________')])
    d['explanations'] = s2[s2.index('___________')+1:]
    
    return d
```
